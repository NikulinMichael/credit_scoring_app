```python
# Словарь под текущие значения 
# Почему мы вообще переносим числовые признаки в категориальные ?
status_values = {
    1: 'ok',
    2: 'default',
    3: 'unk'
}
```

```python
df.status = df.status.map(status_values)
```

```python
df.head()
```

```python
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}

df.home = df.home.map(home_values)
df.home.head()
```

```python
marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}

df.marital = df.marital.map(marital_values)
df.head()
```

```python
record_values = {
    1: 'no',
    2: 'yes',
    3: 'unk'
}

df.record = df.records.map(record_values)
df.record.head()
```

```python
job_values = {
    1: 'fixed',
    2: 'parttime',
    3: 'freelance',
    4: 'others',
    0: 'unk'
}

df.job = df.job.map(job_values)
df.job.head()
```

```python
df.describe().round()
```

```python
# Выбивающиеся значения в income, assets, debt 9999....
# Заменим на NaN
for c in ['income', 'assets', 'debt']:
    df[c] = df[c].replace(to_replace=99999999, value=np.nan) 
```

```python
df.describe().round()
```
